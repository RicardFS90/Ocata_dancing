```python
total_tracks=sp.artist_albums('spotify:artist:4mmlq16Y7gghdUsgzwj1Sh')['items']
tot_albums=[album['id'] for album in total_tracks]
#Manel
#tot_albums=['spotify:album:5cAaJpYGFuIaZmLigIuWBh','spotify:album:2Y9RTKWqlYcTJjJhqX2rYQ','spotify:album:2tW5O96UN1PtRwqQ641HVl','spotify:album:75OD1SnBNNFYHoMuwc3WYV']
#I
#tot_albums=['spotify:album:7hf4Do8ywyKVJiwQqIexc2','spotify:album:4ApLmAxyvhQi7GGK5CMOgG','spotify:album:2J3tzhwNm2jhGATxQmawOt','spotify:album:1H0Kzu6QdUAuUDMGOIe271']
#sanjosex
#tot_albums=['spotify:album:11LvEOKIH5NMnLjrvsuc2b','spotify:album:2NToY2Ne6qJNslACNUc6wP','spotify:album:644dDZjLaAGBJ9vvAkymRK']
songs_ids = []
songs = []
for album in range(len(tot_albums)):
    tracks_IDs = sp.album_tracks(tot_albums[album])
    songs_ids.extend([song['id'] for song in tracks_IDs['items']])
    songs.extend([song['id'], song['name']] for song in tracks_IDs['items'])
```

```python
features=sp.audio_features(tracks=songs_ids)  
songs_df=pd.DataFrame(features)
songs_names_df=pd.DataFrame(songs)
songs_df = pd.merge(left=songs_names_df, right=songs_df, left_on=0, right_on='id')
songs_df.drop(0,axis=1, inplace=True)
songs_df.rename(columns={1:'name'},inplace=True)
songs_df.drop(['mode','type','uri','track_href','analysis_url'],axis=1, inplace=True)
```

```python
tracks_df=pd.DataFrame(sp.tracks(songs_ids)['tracks'])
tracks_df = tracks_df[['id', 'popularity','album']]
tracks_df = pd.merge(left=songs_df, right=tracks_df, left_on='id', right_on='id')
tracks_df['album'] = tracks_df['album'].apply(lambda x:x.get('name'))
tracks_df[['name', 'album','popularity', 'energy', 'danceability', 'tempo', 'acousticness', 'liveness','valence']].sort_values('popularity', ascending = False).reset_index(drop=True)
```
